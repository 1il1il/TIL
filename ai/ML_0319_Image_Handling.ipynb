```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
color_img = Image.open('../data/img/fruits.jpg')
plt.imshow(color_img)
plt.show()
```

```python
#  3 ndarray 
colar_pixel = np.array(color_img)
print('colar_pixel shape : {}'.format(colar_pixel.shape))

#  
#   3  
#  pixel RGB    R,G,B 
gray_pixel = colar_pixel.copy()

for y in range(gray_pixel.shape[0]):
    for x in range(gray_pixel.shape[1]):
        gray_pixel[y,x] = int(np.mean(gray_pixel[y,x]))

plt.imshow(gray_pixel)
plt.show()
```

```python
#   3   with numpy
test = np.mean(colar_pixel, axis=2, dtype=np.int32, keepdims=True)
test_array = np.concatenate([test,test,test], axis=2)
plt.imshow(test_array)
plt.show()
```

```python
# Gray Scaling - Opencv 
r = 0.2989
g = 0.5870
b = 0.1140
gray = colar_pixel[:, :, 0]*r + colar_pixel[:, :, 1]*g + colar_pixel[:, :, 2]*b
gray = 255 - gray
plt.imshow(gray, cmap='Greys')
plt.show()
```

```python
#   2 
gray_2d_pixel = gray_pixel[:, :, 0]
print(gray_2d_pixel.shape)
plt.imshow(gray_2d_pixel, cmap='gray')
plt.show()

# array   
gray_2d_image = Image.fromarray(gray_2d_pixel)
gray_2d_image.save('../data/img/my_gray_image.jpg')
```
