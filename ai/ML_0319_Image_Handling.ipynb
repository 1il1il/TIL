```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = Image.open('../data/img/justice.jpg')
plt.imshow(img)
plt.show()

pixel = np.array(img)
print('X좌료 : {}, Y좌표 : {}의 pixel의 값 : {}'.format(100,200, pixel[200,100]))

print('이미지의 크기 : {}'.format(img.size))
print('이미지의 shape : {}'.format(pixel.shape))

# 이미지에 대한 기본 처리
# 이미지 객체를 이용하여 여러 처리

# 이미지 저장
img.save('../data/img/my_image.jpg')

# 이미지 잘라내기 (좌상, 우하)
crop_img = img.crop((30,100, 150,330))
plt.imshow(crop_img)
plt.show()

# 이미지 크기 변경 > 튜플로(,) 넣어야한다
resize_img = img.resize((int(img.size[0]/8), int(img.size[1]/8)))
plt.imshow(resize_img)
plt.show()

# 이미지 회전
rotate_img = img.rotate(180)
plt.imshow(rotate_img)
plt.show()
```

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
color_img = Image.open('../data/img/fruits.jpg')
plt.imshow(color_img)
plt.show()
```

```python
# 이미지를 3차원 ndarray로 변환
colar_pixel = np.array(color_img)
print('colar_pixel의 shape : {}'.format(colar_pixel.shape))

# 흑백으로 처리하기
# 흑백 이미지를 3차원 데이터로 표현하기
# 각 pixel의 RGB값의 평균을 구해서 각각의 R,G,B값으로 설정
gray_pixel = colar_pixel.copy()

for y in range(gray_pixel.shape[0]):
    for x in range(gray_pixel.shape[1]):
        gray_pixel[y,x] = int(np.mean(gray_pixel[y,x]))

plt.imshow(gray_pixel)
plt.show()
```

```python
# 흑백 이미지를 3차원 데이터로 표현하기 with numpy
test = np.mean(colar_pixel, axis=2, dtype=np.int32, keepdims=True)
test_array = np.concatenate([test,test,test], axis=2)
plt.imshow(test_array)
plt.show()
```

```python
# Gray Scaling - Opencv의 방식
r = 0.2989
g = 0.5870
b = 0.1140
gray = colar_pixel[:, :, 0]*r + colar_pixel[:, :, 1]*g + colar_pixel[:, :, 2]*b
gray = 255 - gray
plt.imshow(gray, cmap='Greys')
plt.show()
```

```python
# 흑백으로 변환 2차원으로 표현하기
gray_2d_pixel = gray_pixel[:, :, 0]
print(gray_2d_pixel.shape)
plt.imshow(gray_2d_pixel, cmap='gray')
plt.show()

# array를 이미지로 변환하여 저장
gray_2d_image = Image.fromarray(gray_2d_pixel)
gray_2d_image.save('../data/img/my_gray_image.jpg')
```
