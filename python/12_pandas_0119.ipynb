```python
import numpy as np
import pandas as pd

np.random.seed(1)

df = pd.DataFrame(np.random.randint(0,10,(6,4)))

df.index =pd.date_range('20200101', periods=6)
# df.index =pd.date_range('20200101', '20200106')
df.columns = ['A','B','C','D']

display(df)
df['E'] = [7, np.nan, 4, np.nan, 2, np.nan]
display(df)
```

```python
# how = any, nan이 하나라도 있으면 해당 행을 drop()
# how = all, 모든 column의 값이 nan인경우 행을 drop()
# inplace = False로 default
new_df = df.dropna(how='any')
display(new_df)
```

```python
# nan을 지정한 값으로 변경
new_df = df.fillna(value=0)
display(new_df)
```

```python
display(df.isnull())
```

```python
# E columns의 값이 nan인 행의 모든 column의 값
display(df.loc[df['E'].isnull()])
```

```python
display(df.loc[df.isnull()['E'],:])
```

```python
# 중복행 제어
my_dict = {
    'k1' : ['one']*3 + ['two']*4,
    'k2' : [1,1,2,3,3,4,4]
}
df = pd.DataFrame(my_dict)
display(df)

# 각행에 대한 중복 여부 확인
print(df.duplicated())
# 중복된 행만 출력
df.loc[df.duplicated(),:]
```

```python
# 중복된 행 제거
display(df.drop_duplicates())
```

```python
# Series, DataFrame의 Grouping
my_dict = {
    '학과' : ['컴퓨터','경영학과','컴퓨터','경영학과','컴퓨터'],
    '학년' : [1, 2, 3, 2, 3],
    '이름' : ['홍길동','신사임당','김연아','아이유','강감찬'],
    '학점' : [1.5, 4.4, 3.7, 4.5, 4.2]
}
df = pd.DataFrame(my_dict)
display(df)
```

```python
# 학과를 기준으로 Grouping
score = df['학점'].groupby(df['학과'])
```

```python
# group안의 데이터 확인하기위해선 get_group()
score.get_group('컴퓨터')
```

```python
# 각 그룹안에 몇개의 데이터가 들어가 있는가
print(score.size())
```

```python
print(score.mean())
```

```python
score1 = df[['학점','이름']].groupby(df['학과'])
display(score1.get_group('컴퓨터'))
print(score1.mean())
```

```python
# 2단계 그룹
score = df['학점'].groupby([df['학과'],df['학년']])
```

```python
# Series가 리턴된다(멀티 인덱스)
print(score.mean())
# 최하위 index를 column으로 바꿔준다
display(score.mean().unstack())
```

```python
# 학과별 평균 학점
print(df['학점'].groupby(df['학과']).mean())

print(df.groupby(df['학과'])['학점'].mean())
```

```python
# 학과별 인원
# 학과로 그룹만들어서 그중 이름 데이터에 대한 count()
print(df.groupby(df['학과'])['이름'].count())
```

```python
print(df.groupby(df['학과']).size())
```

```python
# 그룹핑 이후 반복처리
# 그룹을 만든 기준(dept), 해당 기준으로 만들어진 그룹(group)
for dept, group in df.groupby(df['학과']):
    print(dept)
    display(group)
```
